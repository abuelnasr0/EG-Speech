```python
# now get dataframes for (adapt, dev, test)
import pandas as pd
import re
import os

MGB3_DIR = "/content/drive/MyDrive/vodafone/datasets/mgb3"
transcripts_types = ["adapt", "dev", "test"]
transcriptors = ["Alaa", "Ali", "Mohamed", "Omar"]

dataframe_dict = {
    "adapt": None,
    "dev": None,
    "test": None,
}

def get_df(file, transcriptor):
  values = []
  with open(file, "r") as f:
    skipped=0
    skipped_en=0
    for line in f.readlines():
      line=line[:-1]
      value_dict = dict()
      line_splitted = line.split(" ")

      value_dict["video_id"] = "_".join(line_splitted[:4])

      # Check if the transcript is in Availbe videos
      if value_dict["video_id"] not in AVAILABLE_VIDEOS:
        skipped+=1
        continue

      value_dict["transcript"] = " ".join(line_splitted[8:])

      # Check if text contain English. and skip it.
      if re.search('[a-zA-Z]', value_dict["transcript"]) is not None:
        skipped_en+=1
        continue

      value_dict["start"] = float(line_splitted[4]+"."+line_splitted[5])
      value_dict["end"] = float(line_splitted[6]+"."+line_splitted[7])
      value_dict["transcriptor"] = transcriptor

      values.append(value_dict)


  return pd.DataFrame.from_records(values), skipped, skipped_en

for transcripts_type in transcripts_types:
  type_dir = os.path.join(MGB3_DIR, transcripts_type)
  curr_df = pd.DataFrame(columns=["video_id", "start", "end", "transcript", "transcriptor"])
  for transcriptor in transcriptors:
    transcripts, skipped, skipped_en = get_df(os.path.join(type_dir, transcriptor, "text_noverlap_norm"), transcriptor)
    print("*"*20)
    print(transcripts_type, transcriptor, " | ", skipped, "skipped, ", skipped_en, "English skipped, ", len(transcripts), "remain")
    print("*"*20)


    # Join all dataframes from the same type
    curr_df = pd.concat([curr_df, transcripts], ignore_index=True)
  dataframe_dict[transcripts_type] = curr_df
```

```python
from thefuzz import fuzz
import numpy as np

def get_best_transcript(transcripts):
  sim_ratio = []
  for i in range(len(transcripts)):
    curr_sim_ratio = 0
    for j in range(len(transcripts)):
      if(j==i): continue;
      curr_sim_ratio += fuzz.ratio(transcripts[i], transcripts[j])
    sim_ratio.append(curr_sim_ratio)
  return np.argmax(np.array(sim_ratio))
```

```python
get_best_transcript(all_transcripts_df.iloc[[84, 1031, 1834, 2650], :]["transcript"].values)
```

```python
by_transcript_group = all_transcripts_df.groupby(by=["video_id", "start", "end"])
```

```python
by_transcript_group.indices
```

```python
# Loop over duplicate transcripts
# Choose the best
# drop the rest

all_indexes_to_drop = []
for key in by_transcript_group.indices.keys():
  indexes = by_transcript_group.indices[key]
  best_transcript_index = get_best_transcript(all_transcripts_df.iloc[by_transcript_group.indices[key],:]["transcript"].values)
  indexes_to_drop = np.delete(indexes, best_transcript_index)
  all_indexes_to_drop += indexes_to_drop.tolist()

no_duplicate_df = all_transcripts_df.drop(all_indexes_to_drop)
no_duplicate_df = no_duplicate_df.reset_index()
```

```python
no_duplicate_df = no_duplicate_df.drop(columns=['index'])
no_duplicate_df = no_duplicate_df.reset_index()
no_duplicate_df
```

```python
# Function to cut audios
from pydub import AudioSegment
def trim_audio_and_save(starts, ends, input_file_path, output_dir, output_name):
  # load the audio file
  audio = AudioSegment.from_file(input_file_path)

  outputs_paths = []

  # iterate over the list of time intervals
  for i, (start_time, end_time) in enumerate(zip(starts, ends)):
    # extract the segment of the audio
    segment = audio[start_time*1000:end_time*1000]

    # construct the output file path
    output_file_path_i = os.path.join(output_dir, f"{output_name[i]}.wav")
    outputs_paths.append(output_file_path_i)

    # export the segment to a file
    segment.export(output_file_path_i, format='wav')

for key in transcripts_groups.indices.keys():
  video_id = no_duplicate_df.iloc[transcripts_groups.indices[key], :]["video_id"].values[0]
  audio_path = os.path.join("/content/drive/MyDrive/vodafone/datasets/mgb3/wav", video_id+".wav")
  starts = no_duplicate_df.iloc[transcripts_groups.indices[key], :]["start"].values.tolist()
  ends = no_duplicate_df.iloc[transcripts_groups.indices[key], :]["end"].values.tolist()
  output_names = transcripts_groups.indices[key].tolist()
  trim_audio_and_save(starts, ends, audio_path, TRIMMED_WAV_DIR, output_names)
```
