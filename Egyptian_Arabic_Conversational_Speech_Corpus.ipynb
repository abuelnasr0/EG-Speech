```python
# Function to cut audios
from pydub import AudioSegment
def trim_audio_and_save(intervals, input_file_path, output_file_path):
  # load the audio file
  audio = AudioSegment.from_file(input_file_path)

  outputs_paths = []

  # iterate over the list of time intervals
  for i, (start_time, end_time) in enumerate(intervals):
    # extract the segment of the audio
    segment = audio[start_time*1000:end_time*1000]

    # construct the output file path
    output_file_path_i = f"{output_file_path}_{i}.wav"
    outputs_paths.append(output_file_path_i)

    # export the segment to a file
    segment.export(output_file_path_i, format='wav')

  return outputs_paths
```

```python
# Function to add usefull columns to the

# Function to add column for integer interval instead of str
def convert_time_to_float(x):
  ts, te = x["time"][1:-1].split(",")
  x["interval"] = [float(ts), float(te)]
  return x

# Function for untranscripted audio intrvals. e.g. [*], [UNK]
def check_if_transcripted(x):
  for symbol in UNTRASCRIPTED_SYMBOLS.keys():
    if symbol in x["transcript"]:
      x["untranscripted"] = True
      return x
  x["untranscripted"] = False
  return x

# Function to add a column for topic.
def add_topic(x, audio_id):
  audio_name = audio_id+".wav"
  try:
    topic = AUDIO_INFO[AUDIO_INFO["UTTRANS_ID"] == audio_name]["TOPIC"].values[0]
    x['topic'] = topic
  except IndexError:
    x['topic'] = None
  return x
```

```python
ALL_AUDIOS_DF = pd.DataFrame()

count = 0

for audio_id in AUDIO_IDS:
  # Load audio transcripts into dataframe
  transcripts = pd.read_csv(os.path.join(TRANSCRIPT_DIR, audio_id+".txt"), sep="	", names=TRANSCRIPT_COLUMNS, header=None)

  # Convert time starts and ends into float
  transcripts = transcripts.apply(convert_time_to_float, axis=1)
  transcripts = transcripts.drop("time", axis=1)
  # Untranscripted audios column
  transcripts = transcripts.apply(check_if_transcripted, axis=1)
  # Topic column
  transcripts = transcripts.apply(add_topic, axis=1, audio_id=audio_id)


  # Cut the audio based on the time interval
  source_audio_path = os.path.join(WAV_DIR, audio_id+".wav")
  destination_path =  os.path.join(HF_DATA_DIR, audio_id)
  intervals = transcripts["interval"]
  audio_paths = trim_audio_and_save(intervals, source_audio_path, destination_path)
  count += len(audio_paths)

  # Add audio paths to the dataframe
  transcripts["audio"] = audio_paths

  # Add all audios to one dataframe
  ALL_AUDIOS_DF = pd.concat([ALL_AUDIOS_DF, transcripts])

assert len(ALL_AUDIOS_DF) == count

```
