```python
import pandas as pd
audio_data = pd.read_csv("/content/Arabic_speech_code_switching/DACS_word_level.feat", sep=" ")
```

```python
audio_data
```

```python
audio_data = audio_data.drop(columns=["word_conf", "phone", "phone_conf", "phone_start", "phone_duration", "phone_end", "label1", "label2", "label3"])
audio_data
```

```python
audio_data = audio_data.drop_duplicates()
```

```python
audio_data = audio_data.reset_index()
audio_data = audio_data.drop(columns=["index"])
audio_data
```

```python
from arabic_buckwalter_transliteration.transliteration import buckwalter_to_arabic

def get_arabic_word(x):
  x["AR_word"] = buckwalter_to_arabic(x["word"])
  return x

audio_data = audio_data.apply(get_arabic_word, axis=1)
audio_data
```

```python
def concat_wrods(words):
  return " ".join(words)

transcripts = audio_data.groupby("#id")["AR_word"].agg(concat_wrods).reset_index()
transcripts = transcripts.rename(columns={"AR_word": "transcript"})
transcripts
```

```python
transcripts.iloc[1,:]["transcript"]
```

```python
# Get Audio id and Starts - ends from segments-dacs
audio_info = pd.read_csv("/content/Arabic_speech_code_switching/segments_dacs", sep=" ")
audio_info = audio_info.rename(columns={"#audio_id": "audio_id", "#start": "start", "#end": "end",})

audio_info
```

```python
# Add start and end columns to the transcripts dataframe
transcript_to_join = transcripts.set_index("#id")
audio_info_to_join = audio_info.set_index("#segment_id")

transcripts = pd.concat([transcript_to_join, audio_info_to_join], axis=1, join="inner")
transcripts = transcripts.reset_index()
transcripts = transcripts.rename(columns={"index": "id"})

transcripts
```
